### 2. Marimekko Chart

- Las 5 canciones más reproducidas en Spotify 2023

```python
# Carga de datos
spotify_df = pd.read_csv("C:/Users/jmarc/Desktop/UOC Master/Visualización de datos/PEC 2/Marimekko Chart/spotify-2023.csv", sep=',', encoding='ISO-8859-1')

spotify_df
```

```python
# Se convierte la columna 'streams' a un tipo numérico (float o int)
spotify_df['streams'] = pd.to_numeric(spotify_df['streams'], errors='coerce')

# Si hay valores no numéricos, los convertirá en nulos
spotify_df.dropna(subset=['streams'], inplace=True)

# Selección de las 5 canciones más escuchadas
top_songs = spotify_df.nlargest(5, 'streams')

# Calculo de la proporción del total de streams
total_streams = top_songs['streams'].sum()
top_songs['proportion'] = top_songs['streams'] / total_streams * 100

# Preparación de los datos para el gráfico Marimekko
fig = go.Figure()

# Barra para cada canción
for index, row in top_songs.iterrows():
    fig.add_trace(go.Bar(
        x=[row['proportion']],
        y=[100],  # La altura de todas las barras es la misma
        width=[row['proportion']],
        name=f"{row['track_name']} by {row['artist(s)_name']}",
        text=row['track_name'],
        textposition='inside'
    ))

# Eje X para que las barras se ubiquen correctamente
fig.update_xaxes(
    tickvals=top_songs['proportion'].cumsum() - top_songs['proportion'] / 2,
    ticktext=top_songs['track_name']
)

# Eje Y para que no muestre los valores
fig.update_yaxes(showticklabels=False)

# Actualiza el layout del gráfico
fig.update_layout(
    barmode='overlay',
    title_text='Las 5 canciones más reproducidas en Spotify 2023',
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)

# Se nuestra el gráfico
fig.show()
```
